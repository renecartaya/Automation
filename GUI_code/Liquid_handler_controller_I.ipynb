```python
# ============================================ Packages Loading ============================================#

import serial # require pip install PySerial for Serial communication
from tkinter import *
import tkinter as tk 
import time   #for delay functions
import itertools # used to control the number of steps

# ============================================ GUI Windown logo and name configuration ============================================ #

root = Tk()
root.iconbitmap('C:/Users/renea/Pictures/Logos/CMF_logo.ico') #
root.title('CMF Garching - Liquid handler Control (DEMO)')
root.resizable(width=False, height=False)
root.geometry("460x160")

# ============================================ Frame Configuration ============================================ #

mainFrame = Frame(root, bd=10, bg='gainsboro',relief=RIDGE)
mainFrame.grid()

innerFrame = Frame(mainFrame, bd=5, bg='gainsboro', relief=RIDGE)
innerFrame.grid()

# ============================================ Serial port configuration ============================================ #

ser = serial.Serial('COM6', 9600)  # open serial port
time.sleep(2)                      # wait 2 seconds

# ============================================ Parameters entry ============================================ #

entry_num_rev = Entry(root,width=5)
entry_num_rev.insert(0,"1")
entry_num_revLabel = tk.Label(root,text="Write the number of revolutions")
entry_num_revLabel.grid(row=2,column =0)
entry_num_rev.grid(row=2,column = 1)





# ============================================ Main functions set ============================================ #
 
def Turn_left():
    rev_left = entry_num_rev.get()
    rev_left = int(rev_left)
    for steps in itertools.repeat(None,rev_left):
        ser.write(b'U')
    
def Turn_right(): 
    rev_right = entry_num_rev.get()
    rev_right = int(rev_right)
    for steps in itertools.repeat(None,rev_right):
        ser.write(b'D')
                
def Turn_right_once(): 
    ser.write(b'D')
        
def Turn_left_once():
    ser.write(b'U')

# ============================================ Secundary functions set ============================================ #    
    
def userDataCheck(userInput):
    try:
        int(userInput)
        return True
    except:
        messagebox.showerror("Error", "Please enter a valid integer", icon = 'error')
        return False

def menuBlinkEnable():
    if blinkState.get() != 1:
        blinkState.set(1)
    blinkLED()

def menuTurnOn():
    if blinkState.get() == 1:
        blinkState.set(0)
    ser.write(b'H')

def menuTurnOff():
    if blinkState.get() == 1:
        blinkState.set(0)
    ser.write(b'L')

def menuSave():
    print("Selected Save")

def exitGUI():
    root.destroy()
    
def turn_OFF():
    root.destroy()
    ser.write(b'D')
    
# ============================================ Buttons Widget ============================================ #

btn_On = Button(root, text="<< Turb Left", command=Turn_left)
btn_On.grid(row=0,column = 0)

btn_Off = Button(root, text="Turb Right >>", command=Turn_right)
btn_Off.grid(row=0,column = 1)

btn_Off = tk.Button(root, text="Stop", command=turn_OFF)
btn_Off.grid(row=0, column=5)

blinkState = IntVar()
chkBtn_Blink = tk.Checkbutton(root, text = "One revolution to the right",variable = blinkState, command = Turn_right_once)
chkBtn_Blink.grid(row=10, column = 0)

blinkState = IntVar()
chkBtn_Blink = tk.Checkbutton(root, text = "One revolution to the right",variable = blinkState, command = Turn_left_once)
chkBtn_Blink.grid(row=10, column = 1)

# ============================================ Menu ============================================ #

menuBar = Menu(root)
root.config(menu = menuBar)

fileMenu = Menu(menuBar,tearoff = 0)
menuBar.add_cascade(label ='Precission', menu = fileMenu)  
fileMenu.add_command(label ='High')#, command = menuSave)
fileMenu.add_command(label ='Low')#, command = exitGUI)

settings = Menu(menuBar, tearoff = 0)
menuBar.add_cascade(label ='Velocity', menu = settings) 
settings.add_command(label ='1 RPM')#, command = menuBlinkEnable) 
settings.add_command(label ='5 RPM')#, command = menuTurnOn)
settings.add_command(label ='10 RPM')#, command = menuTurnOff)
settings.add_command(label ='20 RPM')#, command = menuTurnOff)
settings.add_command(label ='30 RPM')#, command = menuTurnOff)
settings.add_command(label ='40 RPM')#, command = menuTurnOff)
settings.add_command(label ='50 RPM')#, command = menuTurnOff)

root.mainloop()
ser.close() # Close properly the serial communication and avoid the need to turn off/on the microcontroller board or to have to unplug/plug again
```
