```python
#############################################################################################################################
#
#                        Demo Para probar control ESC de motor Brushless
#
#############################################################################################################################
############################# Packages loading ###################################

import serial 
from tkinter import *
import tkinter as tk 
import time   #for delay functions

root = Tk()
root.wm_title('ESC Test system')
root.configure(background = '#e0e0e0')
root.iconbitmap('C:/Users/ASUS/Pictures/BlueOrbit-logo/generic-logo.ico')
root.resizable(width=False, height=False)
root.geometry("370x390")

mainFrame = Frame(root, bd=10, bg='gainsboro',relief=RIDGE)
mainFrame.grid()

innerFrame = Frame(mainFrame, bd=5, bg='gainsboro', relief=RIDGE)
innerFrame.grid()

       # ##########################  Serial port configuration ###########################

ser = serial.Serial('COM6', 9600)  # open serial port
time.sleep(2)                      # wait 2 seconds

       # ########################## Functions set ###########################

def turn_LED_ON():
    ser.write(b'H')

def turn_LED_OFF(): 
    ser.write(b'L')

def Stop(): 
    ser.write(b'a')
    
def Speed_1(): 
    ser.write(b'b')
    
def Speed_2(): 
    ser.write(b'c')

def Speed_3(): 
    ser.write(b'd')
    
def Speed_4(): 
    ser.write(b'e')
    
def Speed_5(): 
    ser.write(b'f')

    
        ###########################Buttons Widget###########################

led_On = Button(innerFrame, text="Test LED ON", width=10, height =2, font=('arial',18,'bold'), bd=7, command=turn_LED_ON)
led_On.grid(row=0,column = 0)

led_Off = Button(innerFrame, text="Test LED OFF", width=10, height =2, font=('arial',18,'bold'), bd=7, command=turn_LED_OFF)
led_Off.grid(row=0,column = 1)

btn_stop = Button(innerFrame, text="Stop", width=10, height =2, font=('arial',18,'bold'), bd=7, command=Stop)
btn_stop.grid(row=1,column = 0)

btn_spd1 = Button(innerFrame, text="Speed 1", width=10, height =2, font=('arial',18,'bold'), bd=7, command=Speed_1)
btn_spd1.grid(row=1,column = 1)

btn_spd2 = Button(innerFrame, text="Speed 2", width=10, height =2, font=('arial',18,'bold'), bd=7, command=Speed_2)
btn_spd2.grid(row=2,column = 0)

btn_spd3 = Button(innerFrame, text="Speed 3", width=10, height =2, font=('arial',18,'bold'), bd=7, command=Speed_3)
btn_spd3.grid(row=2,column = 1)

btn_spd2 = Button(innerFrame, text="Speed 4", width=10, height =2, font=('arial',18,'bold'), bd=7, command=Speed_4)
btn_spd2.grid(row=3,column = 0)

btn_spd3 = Button(innerFrame, text="Speed 5", width=10, height =2, font=('arial',18,'bold'), bd=7, command=Speed_5)
btn_spd3.grid(row=3,column = 1)

root.mainloop()
ser.close() # Close properly the serial communication and avoid the need to turn off/on the microcontroller board or to have to unplug/plug again
```
