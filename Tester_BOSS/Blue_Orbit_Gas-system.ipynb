```python
#############################################################################################################################
#
#                        Demo Para probar la imagen del GUI LED
#
#############################################################################################################################
############################# Packages loading ###################################

import serial as ser # require pip install PySerial for Serial communication
from tkinter import *
import tkinter as tk 
import time   #for delay functions

root = Tk()
root.wm_title('Blue Orbit Space Systems')
root.configure(background = '#e0e0e0')
root.iconbitmap('C:/Users/ASUS/Pictures/BlueOrbit-logo/generic-logo.ico')
root.resizable(width=False, height=False)
root.geometry("370x230")

mainFrame = Frame(root, bd=10, bg='gainsboro',relief=RIDGE)
mainFrame.grid()

innerFrame = Frame(mainFrame, bd=5, bg='gainsboro', relief=RIDGE)
innerFrame.grid()

       # ##########################  Serial port configuration ###########################

ser = serial.Serial('COM6', 9600)  # open serial port
time.sleep(2)                      # wait 2 seconds

       # ########################## Functions set ###########################

def turn_LED_ON():
    ser.write(b'H')

def turn_LED_OFF(): 
    ser.write(b'L')

def Gas_position_1(): 
    ser.write(b'P')
    
def Gas_position_2(): 
    ser.write(b'R')
    
        ###########################Buttons Widget###########################

btn_On = Button(innerFrame, text="LED ON", width=10, height =2, font=('arial',18,'bold'), bd=7, command=turn_LED_ON)
btn_On.grid(row=0,column = 0)

btn_Off = Button(innerFrame, text="LED OFF", width=10, height =2, font=('arial',18,'bold'), bd=7, command=turn_LED_OFF)
btn_Off.grid(row=0,column = 1)

btn_Off = Button(innerFrame, text="Position 1", width=10, height =2, font=('arial',18,'bold'), bd=7, command=Gas_position_1)
btn_Off.grid(row=1,column = 0)

btn_Off = Button(innerFrame, text="Position 2", width=10, height =2, font=('arial',18,'bold'), bd=7, command=Gas_position_2)
btn_Off.grid(row=1,column = 1)


root.mainloop()
ser.close() # Close properly the serial communication and avoid the need to turn off/on the microcontroller board or to have to unplug/plug again
```
