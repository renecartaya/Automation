```python
#############################################################################################################################
#
#                        Demo for Power Switch using Relays
#
#############################################################################################################################
############################# Packages loading ###################################

import serial 
from tkinter import *
import tkinter as tk 
import time   #for delay functions

root = Tk()
root.wm_title('Power control system')
root.configure(background = '#e0e0e0')
root.iconbitmap('C:/Users/ASUS/Pictures/BlueOrbit-logo/generic-logo.ico')
root.resizable(width=False, height=False)
root.geometry("370x390")

mainFrame = Frame(root, bd=10, bg='gainsboro',relief=RIDGE)
mainFrame.grid()

innerFrame = Frame(mainFrame, bd=5, bg='gainsboro', relief=RIDGE)
innerFrame.grid()

       # ##########################  Serial port configuration ###########################

ser = serial.Serial('COM6', 9600)  # open serial port
time.sleep(2)                      # wait 2 seconds

       # ########################## Functions set ###########################

def power_ON_1():
    ser.write(b'A')

def power_OFF_1(): 
    ser.write(b'B')
    
    
def power_ON_2():
    ser.write(b'C')

def power_OFF_2(): 
    ser.write(b'D')

    
def power_ON_3():
    ser.write(b'E')

def power_OFF_3(): 
    ser.write(b'F')

    
def power_ON_4():
    ser.write(b'G')

def power_OFF_4(): 
    ser.write(b'H')    
    
    
        ###########################Buttons Widget###########################

led_On = Button(innerFrame, text="Power ON", width=10, height =2, font=('arial',18,'bold'), bd=7, command=power_ON_1)
led_On.grid(row=0,column = 0)

led_Off = Button(innerFrame, text="Power OFF", width=10, height =2, font=('arial',18,'bold'), bd=7, command=power_OFF_1)
led_Off.grid(row=0,column = 1)


led_On = Button(innerFrame, text="Power ON", width=10, height =2, font=('arial',18,'bold'), bd=7, command=power_ON_2)
led_On.grid(row=1,column = 0)

led_Off = Button(innerFrame, text="Power OFF", width=10, height =2, font=('arial',18,'bold'), bd=7, command=power_OFF_2)
led_Off.grid(row=1,column = 1)


led_On = Button(innerFrame, text="Power ON", width=10, height =2, font=('arial',18,'bold'), bd=7, command=power_ON_3)
led_On.grid(row=2,column = 0)

led_Off = Button(innerFrame, text="Power OFF", width=10, height =2, font=('arial',18,'bold'), bd=7, command=power_OFF_3)
led_Off.grid(row=2,column = 1)


led_On = Button(innerFrame, text="Power ON", width=10, height =2, font=('arial',18,'bold'), bd=7, command=power_ON_4)
led_On.grid(row=3,column = 0)

led_Off = Button(innerFrame, text="Power OFF", width=10, height =2, font=('arial',18,'bold'), bd=7, command=power_OFF_4)
led_Off.grid(row=3,column = 1)

root.mainloop()
ser.close() # Close properly the serial communication and avoid the need to turn off/on the microcontroller board or to have to unplug/plug again
```
